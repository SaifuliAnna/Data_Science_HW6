# 2.13 - Создайте новый DataFrame, где для сгруппированных данных по столбцу "Должность", выполните агрегацию данных с помощью метода agg и найдите минимальное и максимальное значение в столбце "Зарплата.в.месяц"

# 2.14 - Создайте функцию fill_avg_salary которая будет возвращать среднее значение заработной платы в месяц. Используйте ее для метода apply и создайте новый столбик "avg"

```python
def fill_avg_salary(data):
    new_pyt_data = python_data.groupby(['Должность'])[['Зарплата.в.месяц']].agg(['min', 'max'])
    result = (new_pyt_data.sum(axis=1)/2)
    return result
```

```python
pyt_mean = fill_avg_salary(python_data)
```

```python
pyt_mean
```

```python
# pyt_mean[0]
```

```python
new_python_data["avg"] = pyt_mean.apply(lambda x: x)
```
