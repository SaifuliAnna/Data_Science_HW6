# **Вторая часть**

Реализуйте программу которая будет строить узор Серпинского

```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
P_x = [0]
P_y = [0]
```

```python
def next_plt(x, y, num):
    if num == 0:
      P_x.append(x/2.0)
      P_y.append(y/2.0)
    
    elif num == 1:
      P_x.append((x + 1)/2.0)
      P_y.append(y/2.0)

    else:
      P_x.append((x + 0.5)/2.0)
      P_y.append((y + (3**0.5))/2.0)
```

```python
for t in range(500000):
    n = random.randint(0, 2)
    next_plt(P_x[-1], P_y[-1], n)
```

# first

```python
plt.plot(P_x, P_y, '+b', markersize=0.01)
plt.title('узор Серпинского')
```

```python
# plt.show()
```

# second

```python
plt.scatter(P_x, P_y, color='green', s=0.0001)
plt.title('узор Серпинского')
```

# third

```python
fig, ax = plt.subplots()
ax.scatter(P_x, P_y, color='black', s=0.0001)
ax.set_title('узор Серпинского')

fig.set_figwidth(15)     #  ширина и
fig.set_figheight(11)    #  высота "Figure"
```
