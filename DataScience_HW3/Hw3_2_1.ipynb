# **Перша частина, третьої домашки**

```python
plt.figure(figsize=(20,10))
result_mean["2014"].plot.bar(rot=27, color=["b", "r", "y", "g", "c", "m", "k", "#058290", "#659290", "#999290"], title="Средняя рождаемость в 1950 и 2019 годах")
```

```python
result_mean[['1950', '2019']].plot(title="Средняя рождаемость в 1950 и 2019 годах", figsize=(15,10))
```

```python
plt.figure(figsize=(15,10))
a = result_mean[['1950']]
b = result_mean[['2019']]
# plt.plot(a, 'b-.', b, 'g:',linewidth=2, marker="D")
plt.plot(a, 'b-.', marker="D", label='1950')
plt.plot(b, 'g:',linewidth=2, marker="D", label='2019')
plt.title("Средняя рождаемость в 1950 и 2019 годах", fontsize=12)
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(15,15))
data = result_mean[['2019']]
plt.pie(
    data,
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.show()
```
